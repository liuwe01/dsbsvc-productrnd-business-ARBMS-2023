```python
import time
import pandas as pd
import numpy as np
import math

# internal tools
from dsgtools.reporting import make_format
from dsgtools.reporting import TableWriter
from dsgtools.reporting import freq
from dsgtools.reporting import bivariate
```

```python
path = "Analytics/Personal Folders/liuwei01/refresh2101_9999/orig_extract_2018_2019_12mPerf.csv"
orig18_19 = pd.read_adls(path, reader = pd.read_csv, dtype = str)
print(orig18_19.shape) 
```

```python
orig18_19["account_type_reported"] = orig18_19.account_type_reported.str.zfill(3)
```

```python
orig18_19 = orig18_19[(orig18_19["AnalysisCreditBad"].notnull())]
bivariate(orig18_19["ArchiveDate"], orig18_19["AnalysisCreditBad"].astype(float))
```

```python
orig18_19 = orig18_19[~orig18_19["ArchiveDate"].isin(['201909', '201910', '201911', '201912'])]
print(orig18_19.shape)
```

```python
freq("InsufficientBusiness", "InsufficientRep", df = orig18_19, cross = False)
```

```python
orig18_19 = orig18_19[(orig18_19["InsufficientBusiness"] == '0') & (orig18_19["InsufficientRep"] == '0') & (orig18_19["AnalysisCreditBad"].notnull())]
print(orig18_19.shape)
sample18_19 = orig18_19.sample(25_000, random_state=0)
print(sample18_19.shape)
```

```python
freq(sample18_19.AnalysisCreditBad)
```

```python
extract_21 = pd.read_adls('Analytics/Personal Folders/liuwei01/Prj2023/origination_extract_Joe/origination_extract_12M_202101_202112.csv', 
                      reader = pd.read_csv, dtype = str, encoding='iso-8859-1')
print(extract_21.shape)
freq("InsufficientBusiness", "InsufficientRep", "AnalysisCreditBad", df = extract_21, cross = False)
```

```python
extract_21 = extract_21[(extract_21["InsufficientBusiness"] == '0') & (extract_21["InsufficientRep"] == '0') & (extract_21["AnalysisCreditBad"].notnull())]
print(extract_21.shape)
sample_21 = extract_21.sample(25_000, random_state=0)
print(sample_21.shape)
```

```python
sbfe = pd.concat([sample18_19, sample_21], ignore_index = False)
print(sbfe.shape)
```

```python
sbfe = sbfe.reset_index(drop = True)
```

```python
sbfe["unique_id"] = sbfe.index + 1
sbfe["unique_id"].is_unique
```

```python
sbfe.to_adls("Analytics/RnD Projects/Product RnD/Business/_ARMBS_Tickets/2023/AIR_937/sbfe_sample_for_DS_wperf.parquet", overwrite = True)
```
